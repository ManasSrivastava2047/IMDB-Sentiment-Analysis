```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.preprocessing import LabelEncoder
import torch.nn as nn
import torch.optim as optim
```

```python
df=pd.read_csv("IMDB Dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df["review"] = df["review"].str.lower()
df
```

```python
def remove_urls(text):
    text = re.sub(r"http\S+|www\S+","",text)
    return text
    
df["review"] = df["review"].apply(remove_urls)
df
```

```python
def remove_punc(text):
    text = re.sub(r"[^A-Za-z0-9\s]","",text)
    return text
    
df["review"] = df["review"].apply(remove_punc)
df
```

```python
def remove_htmltags(text):
    text = re.sub(r"<.*?>","",text)
    return text
    
df["review"] = df["review"].apply(remove_htmltags)
df
```

```python
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
```

```python
def remove_stopwords(text):
    tokens=word_tokenize(text)
    stop_word=set(stopwords.words("english"))
    for i in tokens:
        if i in stop_word:
            text=text.replace(i,"")
    return text
df["review"]=df["review"].apply(remove_stopwords)
```

```python
df
```

```python
def stemming(text):
    ps=PorterStemmer()
    stem_words=[]
    tokens=word_tokenize(text)
    for i in tokens:
        st=ps.stem(i)
        stem_words.append(st)
    return " ".join(stem_words)
df["review"]=df["review"].apply(stemming)
```

```python
df
```

```python
le=LabelEncoder()
df["sentiment"]=le.fit_transform(df["sentiment"])
df
```

```python
y=df["sentiment"]
y
```

```python
from sklearn.feature_extraction.text import  TfidfVectorizer
tf=TfidfVectorizer(max_features=5000)
X=tf.fit_transform(df["review"])
X
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(X,y,test_size=0.2,random_state=42)
x_train
```

```python
x_test.shape
```

```python
import torch
from torch.utils.data import TensorDataset,DataLoader
x_train=x_train.toarray()
x_test=x_test.toarray()
x_test
```

```python
trainset=TensorDataset(
    torch.from_numpy(x_train).float(),
    torch.from_numpy(y_train.values).float()
)
testset=TensorDataset(
    torch.from_numpy(x_test).float(),
    torch.from_numpy(y_test.values).float()
)
```

```python
train_loader = DataLoader(trainset, shuffle=True, batch_size=64)
test_loader = DataLoader(testset, shuffle=True, batch_size=64)
```

```python
class RNN(nn.Module):
    def __init__(self,input_size,hidden_size=256,num_layers=1):
        super().__init__()
        self.hidden_size=hidden_size
        self.num_layers=num_layers
        self.rnn=nn.RNN(input_size,hidden_size,num_layers,batch_first=True)
        self.fc=nn.Linear(hidden_size,1)
    def forward(self,x):
        h0=torch.zeros(self.num_layers,x.size(0),self.hidden_size)
        out,_=self.rnn(x,h0)
        out=self.fc(out[:,-1,:])
        return out        
```

```python
input_size = x_train.shape[1]
model = RNN(input_size)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters())
```

```python
best_loss = float('inf')
epochs=20
for epoch in range(epochs):

    running_loss = 0

    model.train()

    for xb, yb in train_loader:

        optimizer.zero_grad()

        xb = xb.unsqueeze(1)

        outputs = model(xb)

        outputs = torch.sigmoid(outputs.squeeze())

        loss = criterion(outputs, yb)

        loss.backward()

        optimizer.step()

        running_loss += loss.item()

    avg_loss = running_loss / len(train_loader)

    print(f"Epoch {epoch+1}, Loss = {avg_loss:.4f}")

    if avg_loss < best_loss:

        best_loss = avg_loss

        torch.save(model.state_dict(), "best_model.pth")

        print("Best model saved!")
```

```python
model.eval()
with torch.no_grad():
    correct_vals = 0
    tot_vals = 0 
    for xb, yb in test_loader:
        xb = xb.unsqueeze(1)
        outputs = model(xb)
        predicted = (torch.sigmoid(outputs.squeeze()) > 0.5).float()
        tot_vals += yb.size(0)
        correct_vals += (predicted == yb).sum().item()
    print(f"accuracy = {correct_vals/tot_vals*100}")
```

### LSTM approach

```python
class LSTMModel(nn.Module):

    def __init__(self, input_size, hidden_size=256, num_layers=1):

        super().__init__()

        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.lstm = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True
        )

        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):

        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)

        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)

        out, _ = self.lstm(x, (h0, c0))

        out = self.fc(out[:, -1, :])

        return out
```

```python
input_size = x_train.shape[1]
model = LSTMModel(input_size)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
epochs = 20
```

```python
best_loss = float('inf')
for epoch in range(epochs):
    running_loss = 0
    model.train()
    for xb, yb in train_loader:
        optimizer.zero_grad()
        xb = xb.unsqueeze(1)
        outputs = model(xb)
        outputs = torch.sigmoid(outputs.squeeze())
        loss = criterion(outputs, yb)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    avg_loss = running_loss / len(train_loader)
    print(f"Epoch {epoch+1}/{epochs}, Loss = {avg_loss:.4f}")
    if avg_loss < best_loss:
        best_loss = avg_loss
        torch.save(model.state_dict(), "best_lstm_model.pth")
        print("Best LSTM model saved!")
```

```python
model.load_state_dict(torch.load("best_lstm_model.pth"))
model.eval()
with torch.no_grad():
    correct_vals = 0
    tot_vals = 0
    for xb, yb in test_loader:
        xb = xb.unsqueeze(1)
        outputs = model(xb)
        predicted = (torch.sigmoid(outputs.squeeze()) > 0.5).float()
        tot_vals += yb.size(0)
        correct_vals += (predicted == yb).sum().item()
accuracy = correct_vals / tot_vals * 100
print(f"Test Accuracy = {accuracy:.2f}%")
```
